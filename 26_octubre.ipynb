```python
coseno= np.cos(x)
seno= np.sin(x)
plt.plot(x,coseno, 'h-c')
plt.plot(x,seno)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de seno y coseno")
plt.grid(color='green', linestyle='--', lw=0.4)
plt.show()
```

```python
def grafica(x):
  coseno= np.cos(x)
  seno= np.sin(x)
  plt.plot(x,coseno, 'h-c')
  plt.plot(x,seno)
  plt.xlabel("Eje X")
  plt.ylabel("Eje Y")
  plt.title("Gráfica de seno y coseno")
  plt.grid(color='green', linestyle='--', lw=0.4)
  plt.show()
  return None
grafica(x)
```
