```python
mean_values=SL["No. of confirmed deaths"].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
```

```python
SL.isnull().sum()
```

```python
Dates=SL.index[0:-1:10]
NCC=SL["No. of confirmed cases"][0:-1:10]
NCD=SL["No. of confirmed deaths"][0:-1:10]
```

```python
plt.plot(Dates,NCC,"H-y",ms=5,label="Infectados")
plt.plot(Dates,NCD,"*-g",ms=5,label="Decesos")
plt.grid()
plt.title("Avance del Ã©bola en Sierra Leona")
plt.ylabel("Casos confirmados")
plt.xlabel("Periodo de tiempo")
plt.xticks(rotation=90,fontsize=8)
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
plt.plot(Dates,NCC,"d--m",ms=4,label="Infectados")
plt.bar(Dates,NCC,color="m",alpha=0.2)
plt.plot(Dates,NCD,"d--k",ms=4,label="Decesos")
plt.bar(Dates,NCD,color="k",alpha=0.2)
plt.grid()
plt.title("Avance del Ã©bola en Sierra Leona")
plt.ylabel("Casos confirmados")
plt.xlabel("Periodo de tiempo")
plt.xticks(rotation=90,fontsize=8)
plt.legend()
plt.savefig("Avance_del_Ã©bola_en_Sierra_Leona.png",dpi=300,format="png")
plt.show()
```

```python
PIvsD=100*SL["No. of confirmed deaths"]/SL["No. of confirmed cases"]
SL["PIvsD"]=PIvsD
```

```python
SL.head()
```

```python
recuperados=((100-PIvsD)/100)*SL["No. of confirmed cases"]
SL["recuperados"]=recuperados
SL.head()
```

```python
SL["PR"]=(100-PIvsD)
SL.head()
```

```python
PIvsD_mean=SL["PIvsD"].mean()
PR_mean=SL["PR"].mean()
porcentajes=[PIvsD_mean,PR_mean]
etiquetas=["PIvsD","PR"]
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct="%0.1f%%",colors=["r","m"])
plt.axis("equal")
plt.title("Porcentaje de Recuperados contra Decesos")
plt.show()
```
